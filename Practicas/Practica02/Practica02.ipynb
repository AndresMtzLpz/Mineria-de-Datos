## **Paso 03 - Detección de valores atípicos**

```python
DatosCovid.hist(figsize=(14,14), xrot=45)
plt.show()
```

```python
DatosCovid.describe()
```

```python
variables = ['total_vaccinations', 'people_fully_vaccinated', 'daily_vaccinations', 'people_vaccinated_per_hundred']
for col in variables:
  sns.boxplot(col,data=DatosCovid)
  plt.show()
```

```python
DatosCovid.describe(include='object')
```

```python

for column in DatosCovid.select_dtypes(include='object'):
  plt.figure(figsize=(15, 20), dpi=80)
  if DatosCovid[column].nunique() < 200: sns.countplot(y=column, data=DatosCovid)
  plt.show()
```

```python
for col in DatosCovid.select_dtypes(include='object'):
  if DatosCovid[col].nunique() < 200:
    display(DatosCovid.groupby(col).agg(['mean']))
```
